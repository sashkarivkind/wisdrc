```python
def extract_results_from_log(logfile):
    results = {'trained_amps':[],'evaluated_amps':[],'accuracy':[]}

    re_float=r"(\d*\.*\d+)"
    re_train_acc = r'( categorical_accuracy: )'+re_float
    matches = re.findall(r'(trained amp )(\d+)\s*(evaluated amp )(\d+).*?'+re_train_acc,
                         logfile,re.MULTILINE+re.DOTALL)
    for match in matches:
        results['trained_amps'].append(int(match[1]))
        results['evaluated_amps'].append(int(match[3]))
        results['accuracy'].append(float(match[5]))
    return {kk:np.array(vv) for kk,vv in results.items()}
```
